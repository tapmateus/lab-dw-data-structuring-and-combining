```python
import pandas as pd
df1 = pd.read_csv("file1.csv")
df2 = pd.read_csv("file2.csv")
df3 = pd.read_csv("file3.csv")

## return number of columns and rows for each file
print(df1.shape)
print(df2.shape)
print(df3.shape)

#setting the same index for the three
df1.set_index('Customer',inplace=True)
df2.set_index('Customer',inplace=True)
df3.set_index('Customer',inplace=True)

#merging columns and rows not in commun for both df1 and df2 and creating a new df12
df12 = pd.merge(left = df1,
               right = df2,
               on = 'Customer',
               how = 'outer')

#now with both merged we'll merge the new df12 with the remaining df3
customer_data = pd.merge(left = df12,
                right = df3,
                on = 'Customer',
                how = 'outer')

print('the new dataset had a total of', customer_data.shape[0], "columns and", customer_data.shape[1], "rows. so, yes, the new dataset had the columns and rows from the three original datasets")

df1
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
dfcomplete = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
dfcomplete.columns
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
## grouping total_claim_amount sums for each sales channel and rounding them
dfgroupbycomp = dfcomplete.groupby('sales_channel').agg({'total_claim_amount':'sum'})

#pivoting the values in the previous df
dfpivotcomp = dfgroupbycomp.pivot_table(columns = ['sales_channel'], values=['total_claim_amount'])
for column in dfpivotcomp:
    dfpivotcomp[column] = dfpivotcomp[column].round(2)

dfpivotcomp

```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
#creating groupby to calculate mean for customer_lifetime_value for gender / education
dfcustgroup_edu = dfcomplete.groupby('education').agg({"customer_lifetime_value":"mean"})
dfcustgroup_gender = dfcomplete.groupby('gender').agg({"customer_lifetime_value":"mean"})
#also for both
dfcustgroup = dfcomplete.groupby(['gender','education']).agg({"customer_lifetime_value":"mean"})

#creating the pivots
dfcustgroup_edu.pivot_table(columns=['education'], values=['customer_lifetime_value'])
dfcustgroup_gender.pivot_table(columns=['gender'], values=['customer_lifetime_value'])
#also for both
dfcustgroup.pivot_table(columns=['education','gender'], values=['customer_lifetime_value'])
```

The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*
